In this project, I will build a simple linear regression model to predict ice cream revenue based on temperature.

The main goal is to practice using **Pandas**, **Matplotlib**, and **Scikit-learn** by exploring the following:
- The relationship between temperature and revenue.
- How to split data into a training set and a test set.
- How to build, train, and evaluate a simple machine learning model.

## Step 0: Setup

First, we will import the required Python libraries and load the dataset. We will also take an initial look at the data to understand its structure.

```python
import matplotlib.pyplot as plt 
import pandas as pd 

# Load the dataset and display the first 5 rows
dataset = pd.read_csv('Ice Cream.csv')
print(dataset.head())
```

## Step 1: Data Preparation

Before we can train our model, we need to prepare the data. This involves:
1.  Separating our data into **independent variables** (the feature, `X`) and the **dependent variable** (the target, `y`).
2.  Splitting the dataset into a **training set** and a **test set**. The model will learn from the training set, and we will use the test set to evaluate its performance on data it has never seen before.

```python
# Define features (X) and target (y), then split into training and test sets
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

```

## Step 2: Model Building & Training

Now that we have prepared the data, we can build our linear regression model. We will do the following:
1.  Create an instance of the `LinearRegression` model.
2.  "Fit" or train the model on our **training data** (`X_train`, `y_train`). This is the step where the model learns the relationship.

```python
# Create and train the Linear Regression model
from sklearn.linear_model import LinearRegression

regr = LinearRegression()
regr.fit(X_train, y_train)
```

## Step 3: Evaluation & Visualization

This is the most important step where we evaluate our model's performance. We will visualize the results by plotting the model's prediction line against the actual data points for both the training and test sets.

- **Training Set Plot**: Shows how well the model fits the data it learned from.
- **Test Set Plot**: Shows how well the model can generalize and make predictions on unseen data. A good result here indicates a successful model.

```python
# Visualize the model's performance on the training data
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regr.predict(X_train), color='blue')
plt.title('Ice Cream Revenue vs. Temperature (Training Data)')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Revenue (USD)')
plt.show()
```

```python
# Visualize the model's performance on the test data
plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, regr.predict(X_train), color='blue')
plt.title('Ice Cream Revenue vs. Temperature (Test Data)')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Revenue (USD)')
plt.show()
```

## Step 4: Conclusion

From this analysis, we can conclude:

- There is a **strong, positive linear relationship** between temperature and ice cream revenue. As the temperature increases, sales increase predictably.
- The linear regression model was able to **successfully learn** this relationship from the training data.
- The model performed very well on the **test set**, which indicates that it is good at generalizing and making accurate predictions on new data.
